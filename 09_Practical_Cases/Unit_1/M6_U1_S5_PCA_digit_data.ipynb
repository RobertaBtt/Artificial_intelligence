
# PRINCIPAL COMPONENT ANALYSIS (PCA)

This may seem a bit strange in two dimensions, but projection and dimensionality reduction can
be extremely useful when viewing high-dimensional data. Now letâ€™s look at applying PCA to digit
data.

Import the libraries to be used and load the data set.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA

#load data
digits = load_digits()
X = digits.data
y = digits.target
```

Apply the PCA method to the data set, transforming 64 dimensions into 2 and visualize it.

```python
pca = PCA(2) #project from 64 to 2
Xproj = pca.fit_transform(X)
print(X.shape)
print(Xproj.shape)

plt.scatter(Xproj[:,0], Xproj[:,1], c=y, edgecolor='none', alpha=0.5, cmap=plt.cm.get_cmap('nipy_spectral',10))
plt.colorbar()
plt.show()
```

This gives us an idea of the relationship between the digits. Essentially, we have found the optimal
rotation in a 64-dimensional space that allows us to see the layout of the digits, without reference
to the labels. But how much information have we gotten rid of? We can solve this by looking at
the variance:

Here we see that our two-dimensional projection loses a lot of information (as measured by the
variance explained) and that we would need about 20 components to retain 90% of the variance.
Analyzing the following graph for a high-dimensional data set can help you understand the level
of redundancy present when observed multiple times.

```python
pca = PCA().fit(X)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.show()
```
