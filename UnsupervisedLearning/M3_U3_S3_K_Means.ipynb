# K-Means Clustering

```python
# =============================================================================
# K-Means Clustering
# =============================================================================

# Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# Dataset
dataset = pd.read_csv('movies.csv', encoding='latin-1')
X = dataset[["budget", "gross"]].values
```

```python


# K-Means
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 5, init= 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(X) # fit_predict returns the cluster associated with the datapoint

### View clusters (they are painted one by one)
# With this, what we do is specify that we want to apply this to the points of Cluster 1 (index = 0),
# y for column 0 (that of X values) and 1 (that of y values) of those points
# That is, in X there are two columns, we take the one of X (x [0]) and that of y (X [1]) but applied to the points of each cluster
# (that is, instead of all the rows with [:,] I take only those of a cluster [y_label == i])
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s = 100, c = 'blue', label = 'C1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s = 100, c = 'red', label = 'C2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s = 100, c = 'green', label = 'C3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s = 100, c = 'cyan', label = 'C4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s = 100, c = 'magenta', label = 'C5')

# Plot centroids
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Clusters of Movies')
plt.xlabel('X1: Budget ($)')
plt.ylabel('X2: Revenue ($)')
plt.legend()
plt.show()

```
